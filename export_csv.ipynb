```python
from nltk.corpus import stopwords
from nltk.tokenize.casual import TweetTokenizer
import string
from langdetect import detect

STOP_PREFIXES = ("@", "http", "&amp")

def clean_text(comment):
    try:
        if detect(comment) == "en":

            comment = comment.replace("\n", " ")
            comment = comment.lower()
            words = []
            alcohol_words = ["alcohol", "drink", "drinking"]

            for word in TweetTokenizer().tokenize(comment):
                if word:
                    if word.isalpha() and word not in alcohol_words and word not in list(string.punctuation) and word not in stopwords.words("english") and not word.startswith(STOP_PREFIXES) and word != "â€™":
                        words.append(word)
            
            return (" ").join(words)
        
        else:
            return None
    
    except:
        return None

def detect_lang(text):
    try:
        if detect(text) == "en":
            return text
        else:
            return None
    except:
        return None
```

```python
import pandas as pd

data = pd.read_json("twitter-@alcohol.json", lines=True)
data = data.drop(data[data.user_id == 405259565].index)
data = data.drop(data[data.user_id == 617853906].index)
data["content"] = data["content"].apply(clean_text)
```

```python
data = data.dropna(subset="content")
data = data.loc[data["lang"] == "en"]
```

```python
#updated aggregated sentiment analysis
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

def find_positive(comment):
    return sia.polarity_scores(comment)["pos"]

def find_neutral(comment):
    return sia.polarity_scores(comment)["neu"]

def find_negative(comment):
    return sia.polarity_scores(comment)["neg"]

def find_compound(comment):
    return sia.polarity_scores(comment)["compound"]

df_reddit["positive"] = df_reddit["body"].apply(find_positive)
df_reddit["negative"] = df_reddit["body"].apply(find_negative)
df_reddit["neutral"] = df_reddit["body"].apply(find_neutral)
df_reddit["compound"] = df_reddit["body"].apply(find_compound)
```

```python
print("The mean positive sentiment score: {} ({})".format(round(df_reddit["positive"].mean(), 3), round(df_reddit["positive"].std(), 3)))
print("The mean negative sentiment score: {} ({})".format(round(df_reddit["negative"].mean(), 3), round(df_reddit["negative"].std(), 3)))
print("The mean neutral sentiment score: {} ({})".format(round(df_reddit["neutral"].mean(), 3), round(df_reddit["neutral"].std(), 3)))
print("The mean compound sentiment score: {} ({})".format(round(df_reddit["compound"].mean(), 3), round(df_reddit["c
ompound"].std(), 3)))
```
